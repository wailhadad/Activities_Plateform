```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
array = pd.Series([1,2,3,3,3,3,1,1,2,4,4,5,5,4,5])
array.value_counts().plot(
    kind='pie', subplots=True,
    title='Pie Chart',
    autopct='%1.2f%%',
    figsize=(6,6),
    textprops={'fontsize': 10},
    ylabel='proportion',
)
```
